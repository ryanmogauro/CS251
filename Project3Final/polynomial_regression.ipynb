#### Check overfitting with $p = 7$ polynomial regression model

In the cell below:

1. Fit a polynomial regression model with $p = 7$ on the fit set.

```python
fit = linear_regression.LinearRegression(data_fit)
check = linear_regression.LinearRegression(data_check)

degree = 7
fit.poly_regression("X","Y",degree)
check.poly_regression("X","Y",degree)

y_pred = check.predict()
check.R2 = check.r_squared(y_pred)

fit.scatter("X","Y","X/Y training")
plt.figure()
check.scatter("X","Y","X/Y testing")


print("MSE training:",fit.m_sse)
residuals = check.compute_residuals(y_pred)
m_sse = np.mean(residuals**2)
print("MSE validation:",m_sse)
```
